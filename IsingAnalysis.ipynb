## Ising Model Data Analysis

```python
import numpy as np
import utils
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TKAgg')

import utils as tools
```

```python
# user fix parameters!
data = np.load('measurements_G.npy', allow_pickle=True)
dynamics = 'Glauber'
```

```python
data_item = data.item()

temperature = data_item['temperature']
energy = data_item['energy']
magnetisation = data_item['magnetisation']
c_v = data_item['c_v']
chi = data_item['chi']
sigma_c = data_item['sigma_c']

print(f'observables : {data_item.keys()}')
print(f'expected energy : {energy}')
print(f'expected magnetisation : {magnetisation}')
print(f'susceptibility : {chi}')
```

```python
%matplotlib inline
# Create subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize=(10,10), sharex=True)  

# First plot (top)
ax[0].plot(temperature, energy, label="Energy", color="blue")
ax[0].set_ylabel("Energy")
ax[0].set_title(f"Energy against Temperature ({dynamics})")
ax[0].legend(loc='best')
ax[0].grid()

# Second plot (bottom)
ax[1].plot(temperature, magnetisation, label="Magnetisation", color="red")
ax[1].set_xlabel("Temperature")
ax[1].set_ylabel("Magnetisation")
ax[1].set_title(f"Magnetisation against Temperature ({dynamics})")
ax[1].legend(loc='best')
ax[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Create subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize=(10,10), sharex=True)  

# First plot (top)
ax[0].errorbar(temperature, c_v, yerr= (sigma_c)**(1/2), label="C_v", fmt='o', ecolor='black', capsize=5, color="blue")
ax[0].set_ylabel("Heat Capacity")
ax[0].set_title(f"Heat Capacity against Temperature ({dynamics})")
ax[0].legend(loc='best')
ax[0].grid()

# Second plot (bottom)
ax[1].scatter(temperature, chi, label="Susceptibility", color="red")
ax[1].set_xlabel("Temperature")
ax[1].set_ylabel("Susceptibility")
ax[1].set_title(f"Susceptibility against Temperature ({dynamics})")
ax[1].legend(loc='best')
ax[1].grid()

plt.tight_layout()
plt.show()
```
